```python
testd = []
for i , row in test_data.iterrows():
    testd.append(row.values.tolist())
```

```python
testd = np.array(testd)
```

```python
test_labels = np.array(test_labels)
```

```python
train_labels = np.array(train_labels)
```

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.datasets import mnist
from tensorflow.keras import backend as K

import matplotlib.pyplot as plt
import numpy as np
```

```python
traind=traind.reshape((traind.shape[0], -1))
testd=testd.reshape((testd.shape[0], -1))
traind=traind/255.0
testd=testd/255.0
traind.shape
```

```python
lb=LabelBinarizer()

train_labels=lb.fit_transform(train_labels)
test_labels=lb.transform(test_labels)
```

```python
# initialise model

model=Sequential()
model.add(Dense(128, input_shape=(784,),activation="sigmoid"))
model.add(Dense(64, activation="sigmoid"))
model.add(Dense(10, activation="softmax"))
```

```python
sgd=SGD(0.01)
epochs=11
model.compile(loss="categorical_crossentropy", optimizer=sgd, metrics=["accuracy"])
H=model.fit(traind, train_labels, validation_data=(testd, test_labels), epochs=epochs, batch_size=128)
```

```python
H.history
```

```python
plt.figure()
plt.plot(np.arange(0,epochs), H.history['loss'], label='training_loss')
plt.plot(np.arange(0,epochs), H.history['accuracy'], label='training_accuracy')
plt.legend()
plt.show()
```
