```python
#importing Libraries

import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from keras.models import Sequential
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

### Loading the data and preprocessing

```python
train_data = pd.read_csv("./train_data.csv")
test_data = pd.read_csv("./test_data.csv")
```

```python
train_data
```

```python
train_labels = train_data['label']
train_data = train_data.drop(columns='label', axis=1)
train_data
```

```python
traind = []
for i,row in train_data.iterrows():
    traind.append(row.values.tolist())
```

```python
traind = np.array(traind)
train_labels = np.array(train_labels)
```

```python
test_labels = test_data['label']
test_data = test_data.drop(columns='label', axis=1)
test_data
```

```python
testd = []
for i,row in test_data.iterrows():
    testd.append(row.values.tolist())
```

```python
testd = np.array(testd)
test_labels = np.array(test_labels)
```

### Defining Model Architecture

```python
traind , testd = traind/255, testd/255
```

```python
traind = traind.reshape(-1, 32, 32, 3)
testd = testd.reshape(-1,32,32,3)
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels[0]
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss ='categorical_crossentropy', metrics=['accuracy'])
```

```python
H = model.fit(traind, train_labels, epochs=10, validation_data=(testd, test_labels) )
```

```python
#Flatten the CIFAR-10 data from (32, 32, 3) to (3072,)
traind = traind.reshape(-1, 3072)
testd = testd.reshape(-1, 3072)
```

```python
model2 = Sequential()
model2.add(Dense(32, input_shape=(3072,),  activation='relu'))
model2.add(Dense(64, activation='relu'))
model2.add(Dense(10, activation='softmax'))

model2.summary()
```

```python
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
H2 = model2.fit(traind, train_labels, epochs=3, validation_data=(testd, test_labels))
```
